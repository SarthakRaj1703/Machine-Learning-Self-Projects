Importing libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #can split data into training and testing data
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score #to tell us the accuracy of our data
```

Data Collection and Processing Steps

```python
#to load the csv data to a pandas dataframe
sonar_data = pd.read_csv('/content/Copy of sonar data.csv',header = None)
sonar_data.head()
```

```python
sonar_data.describe()
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

Data Pre-processing


```python
#separating data and labels
#X stores values, Y stores labels
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
print(X,Y)
```

Training and Testing Data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state = 1)
# ---- DESCRIPTION ---- #
# X_train, X_test are the data split of X, and Y_train and Y_test are corresponding labels

# train_test_split syntax
# (X,Y) are the matrices upon which the operations are performed
# test_size determines the ratio of the data allocated for testing and training (here 10%)
# stratify classifies data on the basis of labels (R/M)
# random_state is like a seed for splitting data

print(X.shape, X_train.shape, X_test.shape)
```

Logistic Regression based Model Training

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with the training data
model.fit(X_train, Y_train)
```

Model Evaluation


```python
#accuracy of the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy score on the test data (real test)
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(testing_data_accuracy)
```
